# Introducci√≥n a las estructuras de control

Como muchos de los lenguajes de programaci√≥n existentes, Python ejecuta los programas de forma secuencial o lineal, que consiste en realizar las instrucciones una tras otra en el orden en que aparecen.

:::{note} Estructuras de control
Las **estructuras de control** en Python permiten romper esta ejecuci√≥n secuencial o lineal del programa,  haciendo que el c√≥digo pueda tomar decisiones o repetir acciones, volvi√©ndose mucho m√°s flexible e inteligente.

## Ejecuci√≥n secuencial (por defecto)
Python ejecuta las instrucciones de arriba hacia abajo, una tras otra, siguiendo el orden lineal del c√≥digo. Por ejemplo:

```python
print("Paso 1")
print("Paso 2")
print("Paso 3")
```

Aqu√≠, `Paso 1` se imprime primero, luego `Paso 2`, y finalmente `Paso 3`.

:::{card}
:header: 1. **Estructuras condicionales**

Permiten **tomar decisiones**, ejecutando un bloque de c√≥digo s√≥lo si se cumple una condici√≥n.

Las principales son:

- `if`: ejecuta c√≥digo si la condici√≥n es **verdadera**
- `else`: ejecuta c√≥digo si la condici√≥n **no** se cumple
- `elif`: permite probar m√∫ltiples condiciones en orden

Ejemplo sencillo:

```python
edad = 17
if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
```

Si la edad es 17, imprimir√° "Eres menor de edad"; si fuera 18 o m√°s, imprimir√° "Eres mayor de edad".
:::

:::{card}
:header: 2. **Bucles**

Permiten **repetir acciones** o bloques de c√≥digo varias veces.

Las principales son:

- **`for`**: repite un bloque un n√∫mero determinado de veces o sobre una secuencia
- **`while`**: repite el bloque mientras se cumpla una condici√≥n, es decir, un n√∫mero indefinido de veces

Ejemplo `for`:

```python
for i in range(5):
    print("Repetici√≥n", i+1)
```

Ejemplo `while`:

```python
contador = 0
while contador < 5:
    print("Repetici√≥n", contador + 1)
    contador += 1
```

Ambos imprimir√°n ‚ÄúRepetici√≥n 1‚Äù hasta ‚ÄúRepetici√≥n 5‚Äù.
:::
