# Bucle for

:::{note} Bucle for
Se utiliza para **iterar** (*repetir, recorrer*) sobre una secuencia (como un rango, lista, tupla, diccionario, etc.) y ejecutar un bloque de código para cada elemento en esa secuencia.
:::

> Importante: Permite **repetir** un bloque de código un **número determinado** de veces.

La sintaxis básica del bucle for es la siguiente:
```python
for elemento in secuencia:
    # bloque de código a ejecutar para cada elemento
```

Veamos algún ejemplo:

## **Ejemplo 1**: Utilizando la función `range`

````python
for x in range (1,10):
    print ("Hola")
````

:::{dropdown} Explicación del programa
Cuando se ejecuta el bucle `for`, Python comienza a iterar (recorrer) sobre la secuencia generada por `range(1,10)`.

* En la primera iteración, `x` será `1`, y se ejecutará `print("Hola")`.  
* En la segunda iteración, `x` será `2`, y se ejecutará nuevamente `print("Hola")`.  

Este proceso continúa hasta que `x` alcanza `9`, momento en el cual se ejecuta `print("Hola")` por última vez.

La función `range(a,b)` genera una secuencia de números que van desde `a` hasta `b-1` 
:::

## **Ejemplo 2**: Iterando (recorriendo) una `lista`

¡Atención!: El concepto `lista` se estudiará más adelante

````python
frutas = ["manzana", "banana", "cereza"]

for elemento in frutas:
    print("Me gusta la " + elemento)
````

:::{dropdown} Explicación del programa
La **lista** `frutas` contiene tres **elementos**: "manzana", "banana" y "cereza".

El bucle `for` itera (recorre) sobre cada elemento de la lista.

* En la **primera iteración**, el parámetro `elemento` adopta el valor `manzana`.
    * Se imprime la frase `Me gusta la manzana`.
* En la **segunda iteración**, el parámetro `elemento` adopta el valor `banana`.
    * Se imprime la frase `Me gusta la banana`.
* En la **tercera iteración**, el parámetro `elemento` adopta el valor `creeza`.
    * Se imprime la frase `Me gusta la cereza`.

En cada iteración, la variable `fruta` toma el valor de un elemento de la lista, y ese valor se imprime con print().
:::

> En este caso se ha utilizado como nombre del parámetro `elemento` para su mejor comprensión. Es muy extendidio entre los programadores utilizar en su lugar la letra `i` (o  `j`).

De esta manera quedaría:
```python
for i in frutas
```

## **Ejemplo 3**: Iterando sobre una cadena de texto

````python
for letra in "Esternocleidomastoideo":
    print(letra)
````

:::{dropdown} Explicación del programa
La palabra "Esternocleidomastoideo" es una cadena de texto (string). El bucle `for` itera sobre cada carácter (letra) del string. En cada iteración, la variable `letra` toma el valor de un carácter del string. `print(letra)`  imprime la letra actual.
:::
