### 2. Lenguaje orientado a objetos
- En Python, *todo* es un objeto: n√∫meros, cadenas, listas...
- La Programaci√≥n Orientada a Objetos (POO) te permite organizar el c√≥digo en piezas reutilizables (clases y objetos), promoviendo modularidad y reutilizaci√≥n.

:::{seealso} ¬°Haz click! üëà

Para ilustrar c√≥mo la POO nos ayuda a organizar el caos, imagina un *Juego de Rol* (RPG). En lugar de tener variables sueltas para la vida o el da√±o de cada personaje, creamos un "molde" (clase) y generamos tantos h√©roes o villanos como necesitemos.

Aqu√≠ tienes un ejemplo sencillo:

```python  
class Personaje:
    def __init__(self, nombre, vida, da√±o):
        self.nombre = nombre
        self.vida = vida
        self.da√±o = da√±o

    def atacar(self, objetivo):
        objetivo.vida -= self.da√±o
        print(f"{self.nombre} ataca a {objetivo.nombre} por {self.da√±o} puntos de da√±o.")   

# Crear personajes
heroe = Personaje("H√©roe", 100, 15)         
villano = Personaje("Villano", 80, 10)

# Simular un ataque
heroe.atacar(villano)   
print(f"La vida restante del villano es {villano.vida}.")
```
En este ejemplo, la clase `Personaje` define las caracter√≠sticas y comportamientos comunes de todos los personajes. Luego, creamos instancias espec√≠ficas como `heroe` y `villano`, cada uno con sus propios atributos. Esto hace que el c√≥digo sea m√°s organizado y f√°cil de mantener.
