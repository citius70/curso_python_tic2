## a) Entero (o Integer)

Un **entero** (*integer*, en ingl√©s) es un **n√∫mero completo** (sin un punto decimal). Puede ser positivo o negativo.

En Python, como en otros lenguajes de programaci√≥n, los enteros se indican **sin comillas** y se representan con el tipo `int`. Por ejemplo:

````python
myInteger = 5
myNegativeInteger = -3
````

:::{seealso} ¬°Haz click! üëà
:class: dropdown
Un procedimiento muy utilizado en programaci√≥n es el de **incremento**.

Es una operaci√≥n que **aumenta el valor de una variable en una cantidad espec√≠fica**, que generalmente es 1. Se utiliza, por ejemplo, para almacenar el n√∫mero de veces que se repite un proceso.
:::

````python

contador = 2
contador = contador + 1
print(contador)
# La variable contador termina como 3

# Aunque la manera de incrementar una variable es v√°lida como se muestra anteriormente, 
# en Python suele simplificarse as√≠:
contador = 2
contador += 1
print(contador)
````

## b) Decimal (o Float)
Un **decimal** o flotante (*float*, en ingl√©s) es un **n√∫mero que incluye un punto decimal**. Puede ser positivo o negativo

En Python, como en otros lenguajes de programaci√≥n, los enteros se indican **sin comillas** y se representan con el tipo `float`.
 Por ejemplo:

````python
myFloat = 3.14
myNegativeFloat = -0.01
````

Cabe destacar que el **separador de decimales** es el *punto* en lugar de la coma.

## c) Cadena de texto (o String)
Una **cadena de texto** (o *string*) es una secuencia de **caracteres**, o tambi√©n mas com√∫nmente conocido como un texto. En Python, puedes crear una cadena encerrando caracteres entre **comillas simples** (') o **comillas dobles** ("). Por ejemplo:

````python
myString = '¬°Hola, mundo!'
anotherString = "Python es divertido"
````

## d) Booleano (o Boolean)
Un **booleano** es un tipo que solo puede tener uno de dos valores: *Verdadero* o *Falso*. Los booleanos a menudo se usan en condiciones, como en una declaraci√≥n if. Por ejemplo:

````python
enviado = True
entregado = False
````

Pon atenci√≥n en el hecho de que True y False empiezan con **may√∫sculas**. Esto es distinto en otros lenguajes de programaci√≥n.

## Concatenaci√≥n de cadenas de texto


La **concatenaci√≥n** es la operaci√≥n que nos permite **unir varios fragmentos de texto** (strings) para formar uno solo. En Python, existen dos formas principales de mostrar informaci√≥n combinada, cada una con un comportamiento distinto:

### A. Uni√≥n con el operador suma (`+`)

Es la concatenaci√≥n propiamente dicha. Se utiliza para fusionar **variables y textos** en una sola unidad.

* **Espacios manuales:** El operador `+` une los textos de forma literal, por lo que debes incluir espacios en blanco `" "` si quieres que las palabras no aparezcan pegadas.
* **Obligaci√≥n de tipos:** Solo puedes concatenar texto con texto. Si intentas concatenar un n√∫mero (`int` o `float`), Python dar√° error a menos que lo conviertas antes a texto usando `str()`.


````python
# Definici√≥n de variables
nombre = "Elena"
apellido = "Garc√≠a"
edad = 34

# Concatenaci√≥n con el operador + (texto y variables de texto)
# No requiere conversi√≥n obligatoria
mensaje1 = "Me llamo " + nombre + " " + apellido
print(mensaje1)


# Concatenaci√≥n con el operador + (texto y variable num√©rica)
# Requiere conversi√≥n obligatoria: str(edad)
mensaje2 = "Tengo " + str(edad) + " a√±os"

# Visualizaci√≥n del mensaje guardado
print(mensaje2)
````


### B. Separaci√≥n con comas (`,`)

Esta opci√≥n se utiliza exclusivamente dentro de la funci√≥n `print()` para mostrar varios datos a la vez.

* **Espacios autom√°ticos:** Python a√±ade autom√°ticamente un espacio entre cada elemento separado por una coma.
* **Flexibilidad de tipos:** No es necesario convertir los n√∫meros a texto; la funci√≥n `print()` acepta diferentes tipos de datos mezclados directamente.


````python
# Definici√≥n de variables
nombre = "Elena"
edad = 34

# Impresi√≥n con comas
# El espacio se genera solo y no hace falta str()
print("Mis datos son", nombre, edad)
````


### Cuadro comparativo

| M√©todo | ¬øCrea una variable nueva? | ¬øNecesita `str()` para n√∫meros? | ¬øA√±ade espacios? |
| --- | --- | --- | --- |
| **Operador `+`** | **S√≠**, genera un nuevo texto. | **S√≠**, es obligatorio. | No, hay que sumarlos. |
| **Coma `,`** | **No**, solo sirve para imprimir. | **No**, lo hace autom√°tico. | S√≠, autom√°ticamente. |


````python
a√±os = 34
nombre = "Elena"
mensaje = "Me llamo " + nombre +  " y tengo " + str(a√±os) + " a√±os de edad"
print(mensaje)
````

N√≥tese como tuvimos que forzar a que la variable `a√±os` fuera un String, usando la funci√≥n str. Hablaremos de esto m√°s adelante en la gu√≠a.

## Conversi√≥n entre tipos de datos

En Python, puedes **convertir** valores de un tipo a otro usando funciones especiales llamadas **funciones de conversi√≥n** üîÑ. Las m√°s comunes son:

- `int()` - Convierte a n√∫mero entero
- `float()` - Convierte a n√∫mero decimal  
- `str()` - Convierte a texto
- `bool()` - Convierte a valor l√≥gico

**‚ö†Ô∏è Importante**: No todas las conversiones son posibles. Por ejemplo, no puedes convertir la texto "hola" a un n√∫mero entero.

### Ejemplos pr√°cticos

**Convertir decimal a entero:**

````python
mi_decimal = 3.74
mi_entero = int(mi_decimal)
print(mi_entero)  # Resultado: 3
````

> **Nota:** Cuando conviertes un decimal a entero, Python **no redondea**, sino que **elimina** la parte decimal (trunca).

**Convertir n√∫mero a texto:**

````python
mi_numero = 25
mi_texto = str(mi_numero)
print("El n√∫mero es: " + mi_texto)  # Resultado: El n√∫mero es: 25
````

**Convertir texto a n√∫mero:**

````python
edad_texto = "17"
edad_numero = int(edad_texto)
print(edad_numero + 1)  # Resultado: 18
````

## Casos especiales

**Conversiones que funcionan ‚úÖ:**
- `int("123")` ‚Üí `123`
- `float("3.14")` ‚Üí `3.14`
- `str(42)` ‚Üí `"42"`
- `bool(1)` ‚Üí `True`

**Conversiones que NO funcionan ‚ùå:**
- `int("hola")` ‚Üí Error
- `float("abc")` ‚Üí Error


## ¬øC√≥mo conocer el tipo de datos de una variable?

En Python, si alguna vez te encuentras en una situaci√≥n en la que no est√°s seguro del tipo de una variable o simplemente deseas verificarlo, puedes utilizar la funci√≥n `type()`. Esta funci√≥n devuelve el tipo de la variable que le pases como argumento.

````python
# Comprobar tipos de datos
mi_numero = 42
mi_texto = "Python"
mi_decimal = 2.71
mi_booleano = False

print(type(mi_numero)) 
print(type(mi_texto))   
print(type(mi_decimal))
print(type(mi_booleano))
````
