# Ejercicios: Comentarios en Python
En este notebook practicarás cómo usar comentarios de línea y de bloque en Python para hacer tu código más claro.

## Recordatorio rápido
- **Comentario de línea**: empieza con `#` y Python ignora todo lo que hay detrás.
- **Comentario de bloque**: se escribe entre `"""` o `'''` y suele usarse para explicar un programa, una función o un bloque de código.

:::{exercise}
:label: comentarios-ejercicio-1
:class: ejercicio-practico

**Objetivo:** Usar **comentarios de línea** para explicar una instrucción sencilla.

**Instrucciones:**
1. Escribe una instrucción `print()` que muestre tu nombre y tu curso (por ejemplo, "2º Bachillerato").
2. Añade un **comentario de línea** antes de la instrucción explicando lo que hace el código.
3. Comprueba que solo se muestra el mensaje con tu nombre y curso al ejecutar la celda, y que el comentario no aparece en la salida.



:::

:::{exercise}
:label: comentarios-ejercicio-2
:class: ejercicio-practico

**Objetivo**: Usar un **comentario de bloque** (docstring) para describir un programa.

**Instrucciones**:
1. Escribe un programa que calcule el área de un rectángulo de base 6 y altura 4.
2. Antes del código, añade un comentario de bloque (con """) que explique:
    * Qué problema resuelve el programa.
    * Qué fórmula utiliza.
3. Al final, muestra el área por pantalla con un mensaje claro.

:::

:::{exercise}
:label: comentarios-ejercicio-3
:class: ejercicio-practico

**Objetivo:** Mejorar la claridad de un código usando **comentarios adecuados** y corrigiendo comentarios confusos.

**Instrucciones:**
1. Observa el siguiente código mal comentado:
   ```python
   # resta dos números
   precio = 15
   cantidad = 2
   # aquí dividimos
   total = precio * cantidad
   print(total) # imprime la suma
   ```
2. Reescribe el programa **corrigiendo los comentarios** para que describan correctamente lo que hace cada línea o bloque.
3. **Mantén el mismo código**, solo cambia y mejora los comentarios.

:::
