# Comentarios en Python
En este notebook aprenderás a usar comentarios en Python y practicarás con ejercicios diseñados para afianzar el concepto.

Los **comentarios** sirven para explicar el código y facilitar la colaboración o el repaso. Python tiene:
- Comentarios de línea (con `#`)
- Comentarios de bloque (con triples comillas `'''` o `"""`)

```python
print("Bienvenido a Python")
```

**Escribe tu código y tu comentario aquí:**
:::

````python
# Tu comentario de línea aquí
print("Bienvenido a Python")
````

```python
numero = 7
doble = numero * 2
print("El doble es:", doble)
```

**Escribe el comentario de bloque y el código aquí:**
:::

````python
'''
Tu comentario de bloque aquí (explica el programa)
'''
numero = 7
doble = numero * 2
print("El doble es:", doble)
````

```python
# suma x e y
x = 10
y = 8
# aquí se multiplica
resultado = x + y
print(resultado) # muestra el resultado
```

**Escribe la versión corregida aquí:**
:::

````python
# Definimos dos números para sumar
x = 10
y = 8
# Calculamos la suma de x e y
resultado = x + y
print(resultado) # Imprime el resultado final
````
