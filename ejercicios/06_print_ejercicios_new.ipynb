**Objetivo**: Imprimir múltiples valores usando `print()`.

:::{exercise}
:label: print-ejercicio-3
:class: ejercicio-practico

**Objetivo**: Usar f-strings para imprimir con formato.

**Ejemplo de referencia**:
```
radio = 5
pi = 3.1416
area = pi * radio ** 2
print(f"El área del círculo es: {area:.2f}")
# Resultado: El área del círculo es: 78.54
```

**Instrucciones del ejercicio**:
1. Crea la variable `radio = 7`.
2. Importa el módulo `math` y usa `math.pi` en lugar de la variable `pi` para mayor precisión.
2. Calcula el volumen de una esfera:
    $$V = \frac{4}{3} \pi r^3$$
3. Usa un f-string para mostrar: "El volumen de la esfera con radio [radio] es [volumen] m3" 
    * Expresa el volumen con 2 decimales.
