**Objetivo**: Practicar el uso de la biblioteca `math` para cálculos geométricos de triángulos rectángulos.

> Recuerda el teorema de Pitágoras, que establece que en un triángulo rectángulo, el cuadrado de la hipotenusa es igual a la suma de los cuadrados de los catetos:

$$\text{hipotenusa}^2 = \text{cateto}_a^2 + \text{cateto}_b^2$$

$$\text{hipotenusa} = \sqrt{\text{cateto}_a^2 + \text{cateto}_b^2}$$

**Ejemplo de referencia**:
```python
import math

# Calcular la hipotenusa de un triángulo rectángulo
catetoA = 3
catetoB = 4
hipotenusa = math.sqrt(catetoA**2 + catetoB**2)
print(f"Hipotenusa: {hipotenusa}")  # Resultado: 5.0
```
