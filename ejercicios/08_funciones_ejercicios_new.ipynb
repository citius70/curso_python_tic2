## Ejercicio resuelto


````python


def calcular_area_rectangulo(b, a):
    area = b * a
    print(f"El área del rectángulo es: {area} m2")

calcular_area_rectangulo(4, 5)
calcular_area_rectangulo(3, 4)
calcular_area_rectangulo(4.3, 5.6)
````

::: {dropdown} Explicación del programa

La función

```python
def calcular_area_rectangulo(b, a):
```

* Recibe dos valores:
  `b` = base, `a` = altura.
* Calcula el área multiplicando base × altura.
* Muestra el resultado por pantalla con `print()`.

Llamadas:

```python
calcular_area_rectangulo(4, 5)   # Área = 20
calcular_area_rectangulo(3, 4)   # Área = 12
calcular_area_rectangulo(4.3, 5.6)  # Área = 24.08
```

Cada llamada ejecuta la función con valores distintos y muestra el área correspondiente.
